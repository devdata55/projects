```python
#Changing datatype of Activitydate
df.ActivityDate=pd.to_datetime(df.ActivityDate)
```

```python
df.info()
```

```python
#Making One TotalMinutes column by combining several minutes column

df['TotalMinutes']=df['VeryActiveMinutes'] + df['FairlyActiveMinutes'] + df['LightlyActiveMinutes'] + df['SedentaryMinutes']
```

### Scatter plot b/w TotalDistance and Calories

```python
from matplotlib import style
style.use('ggplot')
plt.figure(figsize=(10,5))
sns.scatterplot(x='Calories',y='TotalSteps',data=df)
```

### Pie chart of TotalMinutes

```python
Minutes=['VeryActiveMinutes','FairlyActiveMinutes','LightlyActiveMinutes','SedentaryMinutes']
AvgMinutes=df[['VeryActiveMinutes','FairlyActiveMinutes','LightlyActiveMinutes','SedentaryMinutes']].mean()
```

```python
plt.figure(figsize=(16,9))
plt.pie(AvgMinutes,labels=Minutes,autopct='%.1f%%')
plt.legend=(Minutes)
plt.title('TotalMinutes')
plt.show()
```

```python
plt.figure(figsize=(10,5))
textp={'fontsize':15}
explode=[0,0,0.1,0]
plt.pie(AvgMinutes,labels=Minutes,autopct='%.1f%%',explode=explode,shadow=True,textprops=textp,radius=2)
plt.legend=(Minutes)
plt.title('TotalMinutes')
plt.show()
```

### Piechart of Inactive Minutes daily

```python
df['Day']=df.ActivityDate.dt.day_name()
```

```python
df.head(2)
```

```python
day=df.Day.unique() 
sedmin=df.groupby('Day').SedentaryMinutes.sum()#daywise sedentaryminutes
explode=[0,0,0,0,0,0.5,0]
plt.pie(sedmin,labels=day,autopct='%0.1f%%',shadow=True,radius=2,textprops=textp,explode=explode)

plt.show()
```

### Piechart for daily calories burn

```python
day=df.Day.unique() 
cal=df.groupby('Day').Calories.sum()#daywise sedentaryminutes
explode=[0,0,0,0,0,0.5,0]
plt.pie(cal,labels=day,autopct='%0.1f%%',shadow=True,radius=2,textprops=textp,explode=explode)

plt.show()
```

### BarGraph of daily Inactive minutes

```python
day=df.Day.unique() 
AvgMinutes=df[['VeryActiveMinutes','FairlyActiveMinutes','LightlyActiveMinutes','SedentaryMinutes']].mean()
VAM=df.groupby('Day').VeryActiveMinutes.sum()
FAM=df.groupby('Day').FairlyActiveMinutes.sum()
LAM=df.groupby('Day').LightlyActiveMinutes.sum()
```

```python
plt.figure(figsize=(10,5))
dayindex=np.arange(len(day))
                
width=0.2
plt.bar(dayindex,VAM,width=0.2,color='#FF4040',label="VeryActiveMinutes")
plt.bar(dayindex+width,FAM,width=0.2,color='#FFB90F',label="FairlyActiveMinutes")
plt.bar(dayindex+width+width,LAM,width=0.2,color='#DC143C',label="LightlyActiveMinutes")
#plt.xlabel('Week days')
plt.ylabel('Minutes')
plt.title('Daily Inactive Minutes')
plt.xticks(dayindex,day)
#plt.legend()

plt.show()
```
